```python
# preorder, VLR
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def preorderTraversal(self, root):
        if not root:
            return []
        q = []
        res = []
        
        while (len(q) != 0 or root):
            if root:
                q.append(root)
                res.append(root.val)
                root = root.left
            else:
                root = q.pop(0).right
        return res
                
```

```python
# inorder, LVR
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def inorderTraversal(self, root):
        if not root:
            return []
        q = []
        res = []
        
        while (len(q) != 0 or root):
            if root:
                q.append(root)
                root = root.left
            else:
                root = q.pop(0)
                res.append(root.val)
                root = root.right
        return res
                
```

```python
# postorder 1, LRV
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def postorderTraversal(self, root):
        if not root:
            return []
        q = []
        res = []
        
        while (len(q) != 0 or root):
            if root:
                q.append(root)
                res.insert(0, root.val)
                root = root.right
            else:
                root = q.pop().left
        return res
```

```python
# postorder 2, LRV
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def postorderTraversal(self, root):
        if not root:
            return []
        q = [root]
        res = []
        
        while (len(q) != 0):
            node = q.pop()
            res.insert(0,node.val)
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)

        return res
```

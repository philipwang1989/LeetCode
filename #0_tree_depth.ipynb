```python
# Maximum Depth of Binary Tree
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None: 
            return 0
        else:
            left_height = self.maxDepth(root.left)
            right_height = self.maxDepth(root.right)
            return max(left_height, right_height) + 1
```

```python
# symmetric tree
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root:
            return True
        q = []
        q.append(root)
        q.append(root)
        while len(q) != 0:
            node1 = q.pop(0)
            node2 = q.pop(0)
            if (node1 != None and node2 == None) or (node1 == None and node2 != None):
                return False
            if node1 != None and node2 != None:
                if node1.val != node2.val:
                    return False
                q.append(node1.left)
                q.append(node2.right)
                q.append(node1.right)
                q.append(node2.left)
        return True
            
```

```python
# path sum
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def hasPathSum(self, root, target):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: bool
        """
        if not root:
            return False
        res = []
        self.helper(root, 0, target, res)
        return any(res)
    def helper(self, root, total, target, res):
        if not root.left and not root.right:
            if total+root.val == target:
                res.append(True)
        if root.left:  
            self.helper(root.left, total + root.val, target, res)
        if root.right:
            self.helper(root.right, total + root.val, target, res)
```

```python
# Construct Binary Tree from Inorder and Postorder Traversal
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: TreeNode
        """
        # inorder: LVR
        # postorder: LRV
        pleft = 0
        ileft = 0
        pright = len(postorder) - 1
        iright = len(inorder) - 1
        root = self.helper(inorder, ileft, iright, postorder, pleft, pright)
        return root
        
    def helper(self,inorder,ileft,iright,postorder,pleft,pright):
        if pleft > pright:
            return None
        if pleft == pright:
            return TreeNode(postorder[pleft])
        root = TreeNode(postorder[pright])
        ind_root_i = inorder.index(postorder[pright])
        left_length = ind_root_i - ileft 
        
        root.left = self.helper(inorder,ileft,ind_root_i-1, postorder, pleft, pleft+left_length-1)
        root.right = self.helper(inorder,ind_root_i+1,iright,postorder,pleft+left_length,pright-1)
        
        return root
```

```python
# Populating Next Right Pointers in Each Node - perfect tree
# level order traversal
"""
# Definition for a Node.
class Node(object):
    def __init__(self, val, left, right, next):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
"""
class Solution(object):
    def connect(self, root):
        """
        :type root: Node
        :rtype: Node
        """
        if not root:
            return 
        q = [root]
        q.append(None)
        while len(q) != 0 and q[0] != None:
            qlen = len(q)
            while qlen != 0:
                node = q.pop(0)
                if node != None:
                    node.next = q[0]
                    if node.left:
                        q.append(node.left)
                    if node.right:
                        q.append(node.right)
                else:
                    if qlen > 0:
                        q.append(None)
                qlen -= 1
        return root
```

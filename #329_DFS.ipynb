```python
# 329. Longest Increasing Path in a Matrix
# DFS with memoization
class Solution(object):
    def __init__(self):
        self.m = 0
        self.n = 0
        # self.move = zip([-1,1,0,0], [0,0,1,-1])
        self.move = [(-1, 0), (1, 0), (0, 1), (0, -1)]
        # the result is not the same when zip the two
        # be careful with ze
    def longestIncreasingPath(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        if not matrix or not matrix[0]:
            return 0
        res = 1
        self.m = len(matrix)
        self.n = len(matrix[0])
        dp = [[0] * self.n for _ in range(self.m)]
        for x in range(self.m):
            for y in range(self.n):
                res = max(res, self.DFS(matrix, dp, x, y))
        return res
    def DFS(self, matrix, dp, x, y):
        if dp[x][y]:
            return dp[x][y]
        mx = 1
        for dx, dy in self.move:
            nx = x + dx
            ny = y + dy
            if self.outBound(nx, ny) or matrix[nx][ny] <= matrix[x][y]:
                continue
            temp = 1 + self.DFS(matrix, dp, nx, ny)
            mx = max(mx, temp)
        dp[x][y] = mx
        return dp[x][y]
        
    def outBound(self, x, y):
        return x < 0 or x >= self.m or y < 0 or y >= self.n
        
nums = (
[
  [3,4,5],
  [3,2,6],
  [2,2,1]
] )
nums = [[1,2]]
Solution().longestIncreasingPath(nums)
```

```python
class Solution(object):
    def longestIncreasingPath(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: int
        """
        if not matrix or not matrix[0]:
            return 0
        m, n = len(matrix), len(matrix[0])
        res = 0
        cache = [[-1] * n for _ in range(m)]
        for i in range(m):
            for j in range(n):
                path = self.dfs(matrix, cache, m, n, i, j)
                res = max(res, path)
        return res
        
    def dfs(self, matrix, cache, m, n, i, j):
        if cache[i][j] != -1:
            return cache[i][j]
        directions = [(-1, 0), (1, 0), (0, 1), (0, -1)]
        res = 1
        for dire in directions:
            x, y = i + dire[0], j + dire[1]
            if x < 0 or x >= m or y < 0 or y >= n or matrix[x][y] <= matrix[i][j]:
                continue
            path = 1 + self.dfs(matrix, cache, m, n, x, y)
            res = max(path, res)
        cache[i][j] = res
        return cache[i][j]
nums = [[1,2]]
Solution().longestIncreasingPath(nums)
```

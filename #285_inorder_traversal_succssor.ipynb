```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
# The successor of a node p is the node with the
# smallest key greater than p.val.
# case 1:
# if right exist for the x node, inorder Successor is 
# right then left to the bottom.

# case 2:
# otherwise it is upper in the tree.

# tree look like DFS, inorder with node.vals

# iteration

class Solution:
    def inorderSuccessor(self, root, p):
        # root: 'TreeNode', p: 'TreeNode') -> 'TreeNode'
        # case 1
        if p.right:
            p = p.right
            while p.left:
                p = p.left
            return p
        # case 2
        stack = []
        while len(stack) != 0 or root:
            if root:
                stack.append(root)
                root = root.left
            else:
                root = stack.pop()
                if root and root == p:
                    if not stack:
                        return None
                    else:
                        return stack[-1]
                root = root.right
        return None
        
```

```python
# inorder traversal
# LVR
# iteration
class Solution:
    def inorder(self, root):
        if not root:
            return []
        q = []
        res = []
        while q or root:
            if root: # work with left
                q.append(root)
                root = root.left
            else:
                root = q.pop()
                res.append(root.val)
                root = root.right
        return res
    
```

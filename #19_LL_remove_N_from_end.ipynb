```python
# 19. Remove Nth Node From End of List
# Given a linked list,
# remove the n-th node
# from the end of list and return its head.

# Example:

# Given linked list: 1->2->3->4->5, and n = 2.

# After removing the second
# node from the end, the linked list becomes 1->2->3->5.
# Note:
# Given n will always be valid.

# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

# METHOD 1: LINEAR SCAN - twice

class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        if not head:
            return
        if n == 0:
            return head

        node = head
        node_len = 0
        while node:
            node = node.next
            node_len += 1
        # boundary case: n = node_len
        if n == node_len:
            head = head.next
            return head
        # when reach count = total_node - n - 1, break the node and link to next.next
        node = head
        count = 0
        while node:
            if count == node_len - n - 1:
                if n == 1:
                    node.next = None
                    return head
                else:
                    node.next = node.next.next
                    return head
            node = node.next
            count += 1
        return head
```

```python
# One pass algo:
# Two pointers:
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: ListNode
        :type n: int
        :rtype: ListNode
        """
        if not head:
            return
        if n == 0:
            return head
        fast = head
        slow = head
        # move fast such that fast and slow are separated by n
        for _ in range(n):
            fast = fast.next
        # print(fast.val,slow.val)
        if not fast:
            return head.next
        while fast.next:
            fast = fast.next
            slow = slow.next
        slow.next = slow.next.next
        return head
```

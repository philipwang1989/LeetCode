```python
# First, we swap the first two nodes in the list,
# i.e. head and head.next;
# Then, we call the function self as swap(head.next.next)
# to swap the rest of the list following the first two nodes.
# Finally, we attach the returned head of the sub-list in
# step (2) with the two nodes swapped in step (1) to form 
# a new linked list.
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if head and head.next:
            tmp = head.next
            head.next = self.swapPairs(tmp.next)
            tmp.next = head
            return tmp
        return head
```

```python
# Pascal's Triangle
# f(i,j)=f(i−1,j−1)+f(i−1,j)
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        if numRows == 0:
            return []
        if numRows == 1:
            return [[1]]
        previous = self.generate(numRows - 1)
        last = previous[-1]
        current = [1] * numRows
        for j in range(1,numRows-1):
            current[j] = last[j - 1] + last[j]
        previous.append(current)
        return previous
    
Solution().generate(5)
```

```python
# Reverse linked list
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None
# iteration first, using stack

class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head:
            return head
        node = head
        q = []
        while node:
            q.append(node)
            node = node.next
        head = node = q.pop()
        while len(q) != 0:
            node.next = q.pop()
            node = node.next
        return head
```

```python
# Reverse linked list
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None
# iteration, direct method

class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if not head:
            return head
        prev = None
        current = head
        
        while current:
            nxt = current.next
            current.next = prev
            prev = current
            current = nxt
        
        head = prev
        return head
            
```

```python
# Reverse linked list
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None
# iteration first, using stack

class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if head == None or head.next == None:
            return head
        p = self.reverseList(head.next)
        head.next.next = head # 1->2->1
        head.next = None # 1, 2->1
        return p
        
```

```python
# Climbing Stairs
# brute force 1
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        count = self.helper(n, 0)
        return count
    def helper(self, n, count):
        if n < 0:
            return count
        if n == 0:
            count += 1
            return count
        one = self.helper(n - 1, count)
        two = self.helper(n - 2, count)
        return one + two 
Solution().climbStairs(5)
```

```python
# Climbing Stairs
# brute force w/ memoization -> takes forever
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        memo = [0] * (n + 1)
        count = self.helper(n, 0, memo)
        return count
    def helper(self, n, count, memo):
        if n < 0:
            return 0
        if n == 0:
            return 1
        if memo[count] < 0:
            return memo[count]
        memo[count] = self.helper(n - 1, count, memo) + self.helper(n - 2, count, memo)
        return memo[count]
Solution().climbStairs(35)
```

```python
# Climbing Stairs
# DP
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n == 1:
            return 1
        dp = [0] * (n + 1)
        dp[1] = 1
        dp[2] = 2
        for i in range(3,n+1):
            dp[i] = dp[i-1] + dp[i-2]
        return dp[n]
Solution().climbStairs(35)    
```

```python
# maximum depth of binary tree
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root == None:
            return 0
        left = self.maxDepth(root.left)
        right = self.maxDepth(root.right)
        return max(left,right) + 1
```

```python
# implement pow of n
class Solution(object):
    def myPow(self, x, n):
        """
        :type x: float
        :type n: int
        :rtype: float
        """
        if n == 0:
            return 1
        if n < 0:
            x = 1/x
            n = -n
        if n%2 == 0:
            return self.myPow(x*x, n/2)
        else:
            return x * self.myPow(x, n-1)
            
Solution().myPow(2,-2)
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        if l1 is None:
            return l2
        if l2 is None:
            return l1
        if l2.val > l1.val:
            l1.next = self.mergeTwoLists(l1.next, l2)
            return l1
        else:
            l2.next = self.mergeTwoLists(l1, l2.next)
            return l2
            
```

```python
# K-th Symbol in Grammar
# row 1: 0
# row 2: 01
# row 3: 0110
# row 4: 01101001
# row 5: 01101001/10010110
# brute force
class Solution(object):
    def kthGrammar(self, N, K):
        """
        :type N: int
        :type K: int
        :rtype: int
        """
        nums = self.helper(N)
#         print(nums)
        return nums[K - 1]
    def helper(self, N):
        if N == 1:
            return [0]
        if N == 2:
            nums = [0,1]
            return nums
        nums = self.helper(N - 1)
        return nums + [1 - x for x in nums]
            
        
Solution().kthGrammar(5,7)
```

```python
# K-th Symbol in Grammar
# binary tree
#        0
#       1 0
#      10 01
#   10 01 01 10
# if K=5 (left node) and N=4->(K=3,N=3)->(K=2,N=2)->(K=1,N=1): return 0

class Solution(object):
    def kthGrammar(self, N, K):
        """
        :type N: int
        :type K: int
        :rtype: int
        """
        if N == 1:
            return 0
        if (K % 2 == 0): # even, right node
            # if returns val of parent node
            # then based on parent val. return val. of child
            if self.kthGrammar(N - 1, K / 2) == 0:
                return 1
            else:
                return 0
        else: # odd, left node
            if self.kthGrammar(N - 1, (K + 1)/2) == 0:
                return 0
            else:
                return 1
Solution().kthGrammar(5,8)
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution(object):
    def generateTrees(self, n):
        """
        :type n: int
        :rtype: List[TreeNode]
        """
        return self.helper(1, n+1)
    def helper(self, beg, end):
        result = []
        # base case
        if beg > end:
            return []
        for i in range(beg, end):
            leftTree = self.helper(beg, i)
            rightTree = self.helper(i+1, end)
            if len(leftTree) == 0:
                leftTree.append(None)
            if len(rightTree) == 0:
                rightTree.append(None)
            for j in range(len(leftTree)):
                for k in range(len(rightTree)):
                    tree = TreeNode(i)
                    tree.left = leftTree[j]
                    tree.right = rightTree[k]
                    result.append(tree)
        return result
```
